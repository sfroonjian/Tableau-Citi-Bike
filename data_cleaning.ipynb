```python
def gender_tables0(monthyear, year, month):
    monthyear = monthyear[monthyear["birth year"].apply(lambda x: x.isnumeric())]
    monthyear['birth year'] = monthyear['birth year'].astype(float)
    monthyear.loc[monthyear["gender"] == 0, "gender"] = "unknown"
    monthyear.loc[monthyear["gender"] == 1, "gender"] = "male"
    monthyear.loc[monthyear["gender"] == 2, "gender"] = "female"
    monthyear_gender_count = monthyear.groupby("gender").count()
    monthyear_gender_count = monthyear_gender_count.rename(columns = {"tripduration":"count"})
    monthyear_gender_count = monthyear_gender_count["count"]
    monthyear_gender_avg = monthyear.groupby("gender").mean()
    monthyear_gender_avg = monthyear_gender_avg.rename(columns = {"tripduration":"average trip duration",
                                                                 "birth year":"average birth year"})
    monthyear_gender_avg = monthyear_gender_avg[["average trip duration", "average birth year"]]
    monthyear_gender_avg["count"] = monthyear_gender_count

    female_percentage = monthyear_gender_avg.iloc[0,2] / monthyear_gender_avg["count"].sum() * 100
    male_percentage = monthyear_gender_avg.iloc[1,2] / monthyear_gender_avg["count"].sum() * 100
    unknown_percentage = monthyear_gender_avg.iloc[2,2] / monthyear_gender_avg["count"].sum() * 100
    monthyear_gender_avg["percentage"] = [female_percentage, male_percentage, unknown_percentage]

    monthyear_gender_avg["year"] = [year, year, year]
    monthyear_gender_avg["month"] = [month, month, month]
    monthyear_genders = monthyear_gender_avg.reset_index()
    
    return monthyear_genders
```
